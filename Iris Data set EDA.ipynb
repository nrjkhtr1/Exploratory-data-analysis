```python
# Need for Cumulative Distribution Function (CDF)
# We can visually see what percentage of versicolor flowers have a
# petal_length of less than 1.6
# How to construct and CDF
# How to read A CDF?
# On differentiating CDF you get PDF and on integrating PDF you get CDF

#Plot CDF of petal_length
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density = True)

pdf = counts/(sum(counts))
print(pdf);      #[0.02 0.02 0.04 0.14 0.26 0.26 0.14 0.08 0.00 0.04] probability density
print(bin_edges) #[1.0 1.09 1.18 1.27 1.36 1.45 1.54 1.63 1.72 1.81 1.9]
print(counts)    #[0.22222222 0.22222222 0.44444444 1.55555556 2.88888889 2.88888889 1.55555556 0.88888889 0.00000000 0.44444444] 

#compute CDF
cdf = np.cumsum(pdf)   # cumulative sum of the PDF gives you cumulative distribution function
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show();
```

```python
# Misclassifcation of error if you use petal length only

# For setosa

counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density = True)
pdf = counts/(sum(counts))

#compute CDF
cdf = np.cumsum(pdf)   # cumulative sum of the PDF gives you cumulative distribution function
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)



# For virginica

counts, bin_edges = np.histogram(iris_virginica['petal_length'], bins = 10, density=True)
pdf = counts/(sum(counts))

#compute CDF
cdf = np.cumsum(pdf) # cumulative sum of the PDF gives you cumulative distribution function
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)



# For versicolor

counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins=10, density=True)
pdf = counts/(sum(counts))

#compute CDF
cdf = np.cumsum(pdf) # cumulative sum of the PDF gives you cumulative distribution function
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

#In the below graph it can be seen that the probability of flower at petal_length = 5 is virginica = 95% while that of versicolor = 5%.(With the help of cumulative distribution function)
```

# Mean, Variance and Standard Deviation #

```python
# In abouve plot we don't know how many flowers have a particular value of petal_length, it may be less or more than 1. Therefore
# we use the modified version of 1D scatter plot, i.e, histogram plot.

sns.FacetGrid(data, hue='species', height=5) \
    .map(sns.distplot, 'petal_length') \
    .add_legend();
plt.show();

# Mean variance and standard deviation
print("Means:")
print(np.mean(iris_setosa["petal_length"]))

# Mean with an outlier
print(np.mean(np.append(iris_setosa["petal_length"], 50)), " <== This mean comes by adding the outlier value of 50"); # outlier means = due to some reason an error occured in reading the 
#dataset(error can be due to wrong indexing or any other reason, in this example 50 is the outlier value which is going to be 
# added in our iris_setosa["petal_length"], now the total number of values = 51 and then we find the mean which comes our to be 2.413)

print(np.mean(iris_virginica["petal_length"]))
print(np.mean(iris_versicolor["petal_length"]))
```

```python
#print the standard deviation
 
print("Standard Deviation")
print(np.std(iris_setosa["petal_length"]))
print(np.std(iris_virginica["petal_length"]))
print(np.std(iris_versicolor["petal_length"]))

# Standard deviation is how much average deviation of points from the mean value. If the standard deviation is small it means spread is small.
# majority of my points lie between (mean-std, mean+std)
```

```python
# Mean , variance, std_dev can be corrupted easiy with help of small number of outliers 
# So for mean we have median

print("Median:")
print(np.median(iris_setosa["petal_length"]))
# with an outlier
print(np.median(np.append(iris_setosa["petal_length"], 50)), " <== with outlier")

print(np.median(iris_virginica["petal_length"]))
print(np.median(iris_versicolor["petal_length"]))

# Why adding the outlier won't change the value of the mediam
# Take an example array (2, 5, 4, 7, 6, 8, 5)
#first arrange in sorted order
# array = (2, 4, 5, 5, 6, 7, 8) median = (n+1)/2 th term if n is odd and ((n/2)th + (n/2)+1 th)/2 th term 
# here median = 5
# Now we add an outlier, therefore array becomes = (2, 5, 4, 7, 6, 8, 5, 52)
# sort the array = (2, 4, 5, 5, 6, 7, 8, 52)
# median = (5+6)/2 = 5.5 which is nearly equal to median with outlier (since this is a small dataset example, we are seeing
# a little more difference in the median of an outlier and without outlier)
```

```python
# percentiles

print("\033[1m90th Percentile:")   # "\033[1m" to print bold text in python 3
print("\033[0m",np.percentile(iris_setosa["petal_length"],90))   # "\033[0m" to again come to non-bold text
print(np.percentile(iris_virginica["petal_length"], 90))
print(np.percentile(iris_versicolor["petal_length"], 90))
```

```python
# quantiles

print(np.percentile(iris_setosa["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(iris_virginica["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(iris_versicolor["petal_length"], np.arange(0, 100, 25)))
```
